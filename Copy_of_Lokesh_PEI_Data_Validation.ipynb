```python
import pandas as pd
import json

# Load data
customer_df = pd.read_excel("Customer.xlsx")
order_df = pd.read_csv("Order (1).csv")

with open("Shipping.json") as f:
    shipping_df = pd.DataFrame(json.load(f))

# ===============================
# 1. Data Exploration
# ===============================
print("Customer Shape:", customer_df.shape)
print("Order Shape:", order_df.shape)
print("Shipping Shape:", shipping_df.shape)

# ===============================
# 2. Schema Validation
# ===============================
print("\nCustomer Columns:", customer_df.columns)
print("Order Columns:", order_df.columns)
print("Shipping Columns:", shipping_df.columns)

# ===============================
# 3. Null Checks
# ===============================
print("\nCustomer Nulls:\n", customer_df.isnull().sum())
print("Order Nulls:\n", order_df.isnull().sum())
print("Shipping Nulls:\n", shipping_df.isnull().sum())

# ===============================
# 4. Duplicate Checks
# ===============================
print("\nDuplicate Customers:", customer_df.duplicated().sum())
print("Duplicate Orders:", order_df.duplicated().sum())
print("Duplicate Shipping:", shipping_df.duplicated().sum())

# ===============================
# 5. Primary Key Validation
# ===============================
print("\nCustomer PK Check:", customer_df["Customer_ID"].nunique() == len(customer_df))
print("Order PK Check:", order_df["Order_ID"].nunique() == len(order_df))
print("Shipping PK Check:", shipping_df["Shipping_ID"].nunique() == len(shipping_df))

# ===============================
# 6. Referential Integrity
# ===============================
invalid_orders = order_df[
    ~order_df["Customer_ID"].isin(customer_df["Customer_ID"])
]

invalid_shipping = shipping_df[
    ~shipping_df["Customer_ID"].isin(customer_df["Customer_ID"])
]

print("\nInvalid Orders:", len(invalid_orders))
print("Invalid Shipping:", len(invalid_shipping))

# ===============================
# 7. Business Validation
# ===============================
print("\nShipping Status Distribution:")
print(shipping_df["Status"].value_counts())

print("\nDelivery %:")
print(shipping_df["Status"].value_counts(normalize=True)*100)
```

```python
merged_df = customer_df.merge(order_df, on='Customer_ID', how='inner')
merged_df = merged_df.merge(shipping_df, on='Customer_ID', how='inner')

print("Merged DataFrame Shape:", merged_df.shape)
print("Merged DataFrame Columns:", merged_df.columns)
display(merged_df.head())
```
